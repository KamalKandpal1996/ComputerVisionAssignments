#### Image Transformation <br>
- A function that takes image as an input performs some operation and retrurns image as an output is known as <b>Transformation Function.</b> <br>
- Transformation operations on an image are generally refered to as <b>geometric transformations</b> applied on an image.

#### Types

<u>affine</u> : Preserves parallelism(Parallel sides remain parallel after transformation).<br>
   1. Translation <br>
   2. Rotation <br>
   3. Resizing <br>

<u>nonaffine</u> :<br>- Does not preserve parallelism.<br>- Also known as homorgaphy or projective transformation.

### 1. TRANSLATION

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image=cv2.imread("dog.jpg")
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
print(image.shape)
```

```python
rows, cols = image.shape[:2]
translation_matrix = np.float32([[1,0,70],[0,1,80]])
image_translation = cv2.warpAffine(image, translation_matrix, (cols,rows))

from matplotlib.pyplot import subplots
fig, ax = subplots(1,2, figsize = (12,20))
ax[0].set_title('original Image')
ax[1].set_title('Translated Image')
ax[0].imshow(image)
ax[1].imshow(image_translation)
```

### 2. ROTATION

```python
rotation_matrix = cv2.getRotationMatrix2D((cols/2,rows/2),30,0.5)    #rotate_matrix = cv2.getRotationMatrix2D(center of rotation, angle, scale)
image_rotation = cv2.warpAffine(image, rotation_matrix, (cols,rows))

from matplotlib.pyplot import subplots
fig, ax = subplots(1,2, figsize = (12,20))
ax[0].set_title('original Image')
ax[1].set_title('Rotated Image')
ax[0].imshow(image)
ax[1].imshow(image_rotation)
```

### 3. RESIZING

##### resized_image = cv2.resize(image, dimensions, interpolation)
- an <b>interpolation function’s</b> goal is to examine neighborhoods of pixels and use these neighborhoods to optically increase or decrease the size of the image without introducing distortions (or at least as few distortions as possible).

```python
img_scaled1 = cv2.resize(image,(600,200),fx=5,fy=5,interpolation=cv2.INTER_LINEAR)
img_scaled2 = cv2.resize(image,None,fx=0.5,fy=0.5,interpolation=cv2.INTER_CUBIC)   # fx , fy resize both axes by half
img_scaled3 = cv2.resize(image,(200,400),interpolation=cv2.INTER_AREA)

from matplotlib.pyplot import subplots
fig, ax = subplots(1,4,figsize = (12,20) )

ax[0].set_title('original Image')
ax[1].set_title('Resize1')
ax[2].set_title('Resize2')
ax[3].set_title('Resize3')
ax[0].imshow(image)
ax[1].imshow(img_scaled1)
ax[2].imshow(img_scaled2)
ax[3].imshow(img_scaled3)
```
