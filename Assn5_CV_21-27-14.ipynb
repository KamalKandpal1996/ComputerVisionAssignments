#### <u>IMAGE THRESHOLDING TECHNIQUES</u> 

<b>Image Thresholding</b> -Thresholding is the process of creating black and white  image out  of a  gray scale or full color image by setting exactly those pixels to white whose value is above a given  threshold,  setting  other  pixels  to  black.

<b>Thresholding Problems </b> : <br>
1. Selecting the correct Threshold Value. <br>
2. We consider only the intensity, not any relationships between the pixels. There is no guarantee that the pixels identified by the thresholding process are contiguous.

<b>Threshold Detection Methods</b> :<br> 
1. Bimodal Histogram Analysis <br>
2. Optimal Thresholding <br>
3. Otsu Thresholding

### IMPORTING THE LIBRARIES

```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
image = cv2.imread('dog.jpg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

### APPLYING THRESHOLDING

```python
thresh = 0
maxValue = 255
th, dst = cv2.threshold(image, thresh, maxValue, cv2.THRESH_BINARY);
cv2.imshow('dog',dst)
cv2.waitKey(0)
```
