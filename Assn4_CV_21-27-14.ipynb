#### EDGE DETECTION TECHNIQUES

- Edges : an edge can be defined as a set of contiguous pixel positions where an abrupt change of intensity (gray or color) values occur <br>
- Edge Detection : Edge detection is an image processing technique for finding the boundaries of objects within images. It works by detecting discontinuities in brightness. <br>
- Edge Detection Operators are of two types: <br>
Gradient – based operator which computes first-order derivations in a digital image like, Sobel operator, Prewitt operator, Robert operator <br>
Gaussian – based operator which computes second-order derivations in a digital image like, Canny edge detector, Laplacian of Gaussian 

### 1. Sobel Edge Detector

```python
# Importing Resources
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('car.jpg')
image2 = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)

gray_image = cv2.imread('car.jpg',0)
gray_image = cv2.cvtColor(gray_image , cv2.COLOR_BGR2RGB)

blur_image = cv2.GaussianBlur(gray_image, (3,3), 0)

from matplotlib.pyplot import subplots

fig, ax = subplots(1,3, figsize = (17,20))

ax[0].set_title('Original')
ax[1].set_title('Grayscale')
ax[2].set_title('With Gaussian Blur')
ax[0].imshow(image2)
ax[1].imshow(gray_image)
ax[2].imshow(blur_image)
```

```python
sobelx = cv2.Sobel(src=blur_image, ddepth=cv2.CV_64F, dx=1, dy=0, ksize=5)
sobely = cv2.Sobel(src=blur_image, ddepth=cv2.CV_64F, dx=0, dy=1, ksize=5)
sobelxy = cv2.Sobel(src=blur_image, ddepth=cv2.CV_64F, dx=1, dy=1, ksize=5)

fig, ax = subplots(1,3, figsize = (17,20))

ax[0].set_title('sobelx')
ax[1].set_title('sobely')
ax[2].set_title('sobelxy')
ax[0].imshow(sobelx)
ax[1].imshow(sobely)
ax[2].imshow(sobelxy)
```

### 2. Canny Edge Detector

```python
canny_image = cv2.Canny(image=blur_image, threshold1=150, threshold2=250)
canny_image = cv2.cvtColor(canny_image, cv2.COLOR_BGR2RGB)
plt.imshow(canny_image)
```
