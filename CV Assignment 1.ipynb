```python
import cv2
```

```python
from matplotlib import pyplot as plt
```

```python
img = cv2.imread('cat_cartoon_images.jpg')
```

### TO OPEN THE IMAGE IN DIFFERENT WINDOW

```python
# cv2.imshow('cat',img)
# cv2.waitKey(0)
```

### TO OPEN THE IMAGE IN THE SAME JUPYTER NOTEBOOK WINDOW (USING MATPLOTLIB)

```python
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)   # openCV follows BGR .... matplotlib follows RGB  ... (so coversion is important!!)
```

```python
plt.imshow(img)
plt.xticks([]),plt.yticks([]) #Hides the graph ticks and x/y axis.    
plt.show()
```

### RESIZING THE IMAGE

```python
width, height = 200,200
img_resized = cv2.resize(img,(width,height))   # width is declared first , height is declared second 

plt.imshow(img_resized)
plt.xticks([]),plt.yticks([]) #Hides the graph ticks and x/y axis.    
plt.show()
```

### GRAYSCALING THE IMAGE

```python
img_gray = cv2.imread('cat_cartoon_images.jpg',0)   # 1st way 
# img_gray = cv2.imread('cat_cartoon_images.jpg',cv2.IMREAD_GRAYSCALE)  #2nd way

# cv2.imshow('CAT_GRAY',img_gray)               ******* Using openCV (opens a separate window) *******
# cv2.waitKey(0)

img_gray = cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB)      # ********* using matplotlib (displayed in same window) **********
plt.imshow(img_gray)
plt.xticks([]), plt.yticks([])
plt.show()
```

### DIMENSIONS OF THE IMAGE 

```python
print('Number of pixels : ', img.size)  # 689x1280x3 = 2645760(pixels)
print('Dimension : ',img.shape)      #  (689,1280,3)  : 689- Height(pixels) , 1280- Width(pixels), 3 referes to 3 channels(BGR)
# img.shape[0] is Height    img.shape[1] is Width     img.shape[2] is 3 channels(BGR)
```

### CROP THE IMAGE

```python
img_cropped = img[0:450,200:1100]   # img[height crop,width crop] -- NO NEED of a separate function
# (image itself is a matrix, we only need to change the dimensions)

# cv2.imshow('CAT_CROPPED',img_cropped)              # ******* Using openCV (opens a separate window) *******
# cv2.waitKey(0)

# img_cropped = cv2.cvtColor(img_cropped, cv2.COLOR_BGR2RGB)      # ********* using matplotlib (displayed in same window) **********
plt.imshow(img_cropped)
plt.xticks([]), plt.yticks([])
plt.show()
```

### FLIP THE IMAGE

```python
img_flip = cv2.flip(img,0)  # 0 for flipping about x-axis(vertically)     # 1 for flipping about y-axis(horizontally)

# cv2.imshow('CAT_FLIPPED',img_flip)
# cv2.waitKey(0)

# img_flip = cv2.cvtColor(img_flip, cv2.COLOR_BGR2RGB)     
plt.imshow(img_flip)
plt.xticks([]), plt.yticks([])
plt.show()
```

### DRAWING SHAPES ON THE IMAGE

```python
# 1.) To draw a line on the image
# img_line = cv2.line(img,(0,0),(250,200),(0,255,0),5) # 2nd and 3rd arguments are the coordinates of start and end point of the line 
# 4th argument is (B,G,R)  #5th argument is thickness of the line

# 2.) To draw a rectangle 
# img_rect = cv2.rectangle(img,(600,600),(100,800),(0,0,255),5)   # Writing -1 instead of 5 for thickness fills the entire rectangle with that color.

# 3.) To draw a circle
img_circle = cv2.circle(img,(640,350),250,(0,255,0),10)   # 2nd argument is coordinates of centre , 3rd argument is the radius

# cv2.imshow('CATCIRCLE',img_circle)
# cv2.waitKey(0)

# img_circle = cv2.cvtColor(img_circle, cv2.COLOR_BGR2RGB)     
plt.imshow(img_circle)
plt.xticks([]), plt.yticks([])
plt.show()
```

### PUTTING TEXT ON THE IMAGE

```python
font = cv2.FONT_HERSHEY_SIMPLEX   #(SELECT FROM AVAILABLE FONTS)
img_text = cv2.putText(img,'HAPPY CAT openCV', (40,650), font, 4, (0,255,0), 10, cv2.LINE_AA )
# 3rd argument - coordinates of bottom-left corner of the text string
# 5th argument - font scale
# 6th argument - BGR
# 7th argument - thickness 
# 8th argument - linetype (optional)

# cv2.imshow('CATFONT',img_text)
# cv2.waitKey(0)

img_text = cv2.cvtColor(img_text, cv2.COLOR_BGR2RGB)     
plt.imshow(img_text)
plt.xticks([]), plt.yticks([])
plt.show()
```
